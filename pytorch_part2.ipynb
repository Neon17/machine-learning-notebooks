## PyTorch tensors and Numpy

NumPy is a popular scientific Python numerical computing library

And because of this, PyTorch has functionality to interact with it

- Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
- PyTorch tensor -> NumPy -> `torch.Tensor.numpy()
- Default Datatype for PyTorch tensors is float32, whereas for NumPy arrays it's float64

```python
# NumPy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) # we can add .type(torch.float32) to change the datatype of the tensor,
# when converting from a NumPy array to a PyTorch tensor, pytorch reflects numpy default datatype of float64
array, tensor
```

```python
array.dtype
```

```python
torch.arange(1.0, 8.0)
```

```python
array = array + 1
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

## Reproductibility (trying to make random out of random)

In short how a neural network work

`start with random numbers -> tensor operations -> update random numbers -> update random numbers to try and make them of the data -> again -> again -> again...`

To reduce the randomness in a neural networks and PyTorch comes the concept of a random seed.

Essentially what the random seed does is "flavour" the randomness

```python
import torch

# Create 2 random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print (random_tensor_A)
print (random_tensor_B)
print (random_tensor_A == random_tensor_B)
```

```python
# Let's make some random but reproducible tensors
import torch

# Set the random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED) # reset the random seed to the same value
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
